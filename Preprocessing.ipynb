```python
import pandas as pd
```

```python
df = pd.read_csv("main_file.csv", header=None)
```

```python
print(df)
```

```python
df.columns = ['Checking Acc Status', 'Duration(months)', 'Credit History', 'Purpose', 'Credit Amount', 'Savings', 'Present employment since', 'Installment rate', 'Personal Status/Gender', 'Gurantors', 'Present residence since', 'Property', 'Age(Years)', 'Other installment plans', 'Housing', 'Number of existing credits at this bank', 'Job', 'People Maintainance', 'Telephone', 'foreign worker', 'Loan Quality']
```

```python
print(df)
```

```python
df.iloc[:,8]
```

```python
symbol_map = {
    'A91': ('male', 'divorced/separated'),
    'A92': ('female', 'divorced/separated/married'),
    'A93': ('male', 'single'),
    'A94': ('male', 'married/widowed'),
    'A95': ('female', 'single')
}
```

```python
def decode_symbol(symbol):
    return symbol_map.get(symbol, (None, None))
df[['Gender', 'Marital_Status']] = df.iloc[:,8].apply(decode_symbol).apply(pd.Series)
```

```python
df = df.drop(columns=['Personal Status/Gender'])
```

```python
df['Loan Quality'] = df['Loan Quality'].map({1: 0, 2: 1})
```

```python
df['Gender'] = df['Gender'].map({'male': 0, 'female': 1}) # male = 0 female = 1
```

```python
df['Marital_Status'] = df['Marital_Status'].map({'single': 0, 'divorced/separated/married': 1, 'divorced/separated': 1, 'married/widowed': 1})
```

```python
df
```

```python
symbol_to_text = {
    # Attribute 1: checking_account
    'A11': '0',
    'A12': '200',
    'A13': '200+',
    'A14': 'none',

    # Attribute 3: credit_history --> ONE HOT
    'A30': 'no credits/all paid',
    'A31': 'all credits paid (this bank)',
    'A32': 'credits paid till now',
    'A33': 'delay in past',
    'A34': 'critical/other credits',

    # Attribute 4: purpose --> ONE HOT
    'A40': 'car (new)',
    'A41': 'car (used)',
    'A42': 'furniture/equipment',
    'A43': 'radio/television',
    'A44': 'domestic appliances',
    'A45': 'repairs',
    'A46': 'education',
    'A47': 'vacation',
    'A48': 'retraining',
    'A49': 'business',
    'A410': 'others',

    # Attribute 6: savings
    'A61': '100',
    'A62': '500',
    'A63': '1000',
    'A64': '1000+',
    'A65': 'unknown',

    # Attribute 7: employment
    'A71': 'unemployed',
    'A72': '1',
    'A73': '4',
    'A74': '7',
    'A75': '7+',

    # Attribute 10: other_debtors --> guarantor and co-applicant are same weight
    'A101': 'none',
    'A102': 'co-applicant',
    'A103': 'guarantor',

    # Attribute 12: property --> ONE HOT
    'A121': 'real estate',
    'A122': 'life insurance/building society',
    'A123': 'car/other',
    'A124': 'unknown',

    # Attribute 14: installment_plans --> ONE HOT
    'A141': 'bank',
    'A142': 'stores',
    'A143': 'none',

    # Attribute 15: housing --> ONE HOT
    'A151': 'rent',
    'A152': 'own',
    'A153': 'for free',

    # Attribute 17: job
    'A171': 'unskilled_non-res',
    'A172': 'unskilled_res',
    'A173': 'skilled',
    'A174': 'management/self-employed',

    # Attribute 19: telephone
    'A191': 0,
    'A192': 1,

    # Attribute 20: foreign_worker
    'A201': 1,
    'A202': 0
}
df_replaced = df.replace(symbol_to_text)
```

```python
pd.options.display.max_columns = None
```

```python
df_replaced
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder(categories=[['unskilled_non-res', 'unskilled_res', 'skilled', 'management/self-employed']])
encoded = encoder.fit_transform(df_replaced[['Job']])
df_replaced['Job'] = encoded
```

```python
df_replaced['Gurantors'] = df_replaced['Gurantors'].map({'none': 0, 'guarantor': 1, 'co-applicant': 1})
```

```python
df_replaced
```

```python
df_final = pd.get_dummies(df_replaced, columns=['Property', 'Other installment plans', 'Housing', 'Credit History', 'Purpose'])
```
